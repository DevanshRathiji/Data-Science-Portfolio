```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

```

```python
path = 'https://raw.githubusercontent.com/amankharwal/Website-data/master/bank.csv'
data = pd.read_csv(path)
data.head()
```

```python
corr_matrix = data.corr().abs()
high_corr_var=np.where(corr_matrix>0.8)
high_corr_var=[(corr_matrix.columns[x],corr_matrix.columns[y]) for x,y in zip(*high_corr_var) if x!=y and x<y]
```

```python
X = data.drop(["Bankrupt?"], axis="columns")
y = data["Bankrupt?"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.feature_selection import SelectKBest,mutual_info_classif
skb =  SelectKBest(mutual_info_classif,k=25)
X_new = skb.fit_transform(X_train,y_train)
cols = skb.get_feature_names_out()
```

```python
X_train,X_test = X_train[cols],X_test[cols]
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_train, y_train)
```

```python
logreg.score(X_test, y_test)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
pipe = make_pipeline(StandardScaler(), LogisticRegression())
pipe.fit(X_train, y_train)  
# apply scaling on training data
```

```python
pipe.score(X_test,y_test)
```
